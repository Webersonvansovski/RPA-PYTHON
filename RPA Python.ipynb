## Desafio

- Vamos automatizar a extração de informações de um sistema e envio de um relatório por e-mail
- No nosso caso, para todo mundo conseguir fazer o mesmo programa, o nosso "sistema" vai ser o Gmail, mas o mesmo processo pode ser feito com qualquer programa do seu computador e qualquer sistema
    - Passo 1: Entrar no sistema (entrar no Gmail)
    - Passo 2: Entrar em uma aba específica do sistema onde tem o nosso relatório (Aba Contatos)
    - Passo 3: Exportar o Relatório (Exportar Contatos)
    - Passo 4: Pegar o relatório exportado, tratar e pegar as informações que queremos
    - Passo 5: Preencher/Atualizar informações do sistema (No nosso caso, criar um e-mail e enviar)

```python
import pyautogui
import time

    # comandos 

# pyautogui.write() -> escreve
# pyautogui.click() -> clica 
# pyautogui.locateOnScreen() -> idenditifa uma imagem na tela
# pyautogui.hotkey() -> usa atalhos do teclado (combinação de teclas)
# pyautogui.press() -> aperta um botão do teclado


    # Lógica / código
pyautogui.alert('O código vai começar. Não mexa em NADA enquanto o código tiver rodando. Quando Finalizar, Você Será Avisado(a)')
    
pyautogui.PAUSE = 1
# abrir o sistema (abrir o navegador = Google Chrome)
# apertar a tecla windows no teclado
pyautogui.press('win')

# digitar Chrome
pyautogui.write('chrome')

# apertar Enter
pyautogui.press('enter')

# Entrar no Gmail
pyautogui.write('Gmail')
pyautogui.press('enter')

time.sleep(2)
# localizar a imagem -> vai ter 4 informações: posicao x, posicao y, largura e altura
x, y, largura, altura = pyautogui.locateOnScreen('busca_gmail.png')
# clicar no meio da imagem
pyautogui.click(x + largura/2, y + altura/2)

# esperar imagem
while not pyautogui.locateOnScreen('logo_gmail.png'):
    time.sleep(1)

# entrar em contatos
x, y, largura, altura = pyautogui.locateOnScreen('pontinhos.png')
pyautogui.click(x + largura/2, y + altura/2)

x, y, largura, altura = pyautogui.locateOnScreen('contatos.png')
pyautogui.click(x + largura/2, y + altura/2)

# esperar contatos
while not pyautogui.locateOnScreen('contatos_espera.png'):
    time.sleep(1)

# exportar os contatos
x, y, largura, altura = pyautogui.locateOnScreen('exportar.png')
pyautogui.click(x + largura/2, y + altura/2)
x, y, largura, altura = pyautogui.locateOnScreen('confirmar_exportar.png')
pyautogui.click(x + largura/2, y + altura/2)
```

### Agora vamos escrever o e-mail

```python
import pandas as pd
import pyperclip

df = pd.read_csv(r'C://users/wvans/Downloads/contacts.csv')
df = df.dropna(axis=1)
display(df)


pyautogui.hotkey('ctrl', 'pgup')

for email in df['E-mail 1 - Value']:
    # clicar no botão de escrever
    while not pyautogui.locateOnScreen('escrever_em.png'):
        time.sleep(1)

    x, y, largura, altura = pyautogui.locateOnScreen('escrever_em.png')
    pyautogui.click(x + largura/2, y + altura/2)
    
    # escrever o email
    pyautogui.write(email)
    pyautogui.press('enter')
    
    # tab para o assunto do e-mail
    pyautogui.press('tab')
    pyautogui.write('weberson caloteiro')
    # tab para o corpo do e-mail
    pyautogui.press('tab')
    texto = '''
    coé Weberson,
    
    para de dar calote e me paga meus 0.20 centavos aé, namoral.
    
    
    att;
    agiota tmj.'''
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.hotkey('ctrl', 'enter')
    
    
pyautogui.alert('O código terminou, pode mexer no seu computador novamente')
```
